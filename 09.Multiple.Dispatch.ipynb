```julia
foo(x, y) = println("duck-typed foo")
foo(x::Int, y::Float64) = println("foo: Int + Float64")
foo(x::Int, y::Int) = println("foo: Two Ints")
```

```julia
foo(1,1)
```

```julia
foo(1.,1.)
```

```julia
foo(1,1.)
```

```julia
foo(true, false)
```
